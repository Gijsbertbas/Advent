```python
nc = 0 # number of children
nm = 0 # number of metadata
meta = 0 # actual metadata
start = True
def fill(header, meta, nc, start):
    print(header)
    if start:
        nc += header[0]
        for i in range(header[1]):
            nc -= 1
            return fill(header[2:], meta, nc, False)
    else:
        while len(header) > 0:
            print(nc)
            if (header[0] == 0) & (nc > 0):
                for i in range(header[1]):
                    meta += header[2+i]
                    print(meta)
                return fill(header[2+header[1]:], meta, nc, False)
            elif nc == 0:
                for h in range(header):
                    meta += h
                    return fill('', meta, nc, False)
            else:
                nc += header[0]
                for i in range(nc):
                    return fill(header[2:], meta, nc, False)
                nc -= 1
    return meta
```

```python
def fill(header, meta=0, nm=0):
    if len(header) == nm:
        for h in header:
            meta += h
    else:
        nc = header[0]
        if nc == 0:
            for i in range(header[1]):
                meta += header[2+i]
            return fill(header[2+header[1]:], meta, nm)
        else:
            nm += header[1]
            for c in range(nc):
                return fill(header[2:], meta, nm)
    return meta
```
