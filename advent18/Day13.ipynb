```python
with open('input/day13.txt') as f:
    start = f.readlines()
startdf = pd.DataFrame(data=[list(s) for s in start]).drop(150, axis=1)
```

```python
def getcarts(startmap):
    carts = []
    rows, cols = startmap.shape
    newmap = startmap.copy()
    for r in range(rows):
        for c in range(cols):
            if newmap.iloc[r,c] == '^':
                carts.append([(r,c),r,c,'up','left'])
                newmap.iloc[r,c] = '|'
            elif newmap.iloc[r,c] == 'v':
                carts.append([(r,c),r,c,'down','left'])
                newmap.iloc[r,c] = '|'
            elif newmap.iloc[r,c] == '>':
                carts.append([(r,c),r,c,'right','left'])
                newmap.iloc[r,c] = '-'
            elif newmap.iloc[r,c] == '<':
                carts.append([(r,c),r,c,'left','left'])
                newmap.iloc[r,c] = '-'

    return newmap, pd.DataFrame(data=carts, columns=['coords','row','column','nextmove','nextturn'])
```

```python
def movecarts(startmap, cartsdf):
    nxttrn = {'right': 'left', 'straight': 'right', 'left': 'straight'}
    trnle = {'up': 'left', 'left': 'down', 'down': 'right', 'right': 'up'}
    trnri = {'up': 'right', 'right': 'down', 'down': 'left', 'left': 'up'}
    for cart in cartsdf.sort_values(['row','column']).iterrows():
        i = cart[0]
        if cart[1]['nextmove'] == 'up':
            cartsdf.loc[i,'row'] -= 1
            newcoords = (cartsdf.loc[i,'row'],cartsdf.loc[i,'column'])
            if newcoords in list(cartsdf['coords'].values):
                print('First crash happend at {}'.format(newcoords))
                return newcoords
            cartsdf.at[i,'coords'] = newcoords
            sign = startmap.iloc[cartsdf.loc[i,'row'],cartsdf.loc[i,'column']]
            if sign == '+':
                if cart[1]['nextturn'] == 'straight':
                    pass
                elif cart[1]['nextturn'] == 'left':
                    cartsdf.loc[i,'nextmove'] = trnle[cart[1]['nextmove']]
                elif cart[1]['nextturn'] == 'right':
                    cartsdf.loc[i,'nextmove'] = trnri[cart[1]['nextmove']]
                cartsdf.loc[i,'nextturn'] = nxttrn[cart[1]['nextturn']]
            elif sign == '/':
                cartsdf.loc[i,'nextmove'] = 'right'
            elif sign == '\\':
                cartsdf.loc[i,'nextmove'] = 'left'
        elif cart[1]['nextmove'] == 'down':
            cartsdf.loc[i,'row'] += 1
            newcoords = (cartsdf.loc[i,'row'],cartsdf.loc[i,'column'])
            if newcoords in list(cartsdf['coords'].values):
                print('First crash happend at {}'.format(newcoords))
                return newcoords
            cartsdf.at[i,'coords'] = newcoords
            sign = startmap.iloc[cartsdf.loc[i,'row'],cartsdf.loc[i,'column']]
            if sign == '+':
                if cart[1]['nextturn'] == 'straight':
                    pass
                elif cart[1]['nextturn'] == 'left':
                    cartsdf.loc[i,'nextmove'] = trnle[cart[1]['nextmove']]
                elif cart[1]['nextturn'] == 'right':
                    cartsdf.loc[i,'nextmove'] = trnri[cart[1]['nextmove']]
                cartsdf.loc[i,'nextturn'] = nxttrn[cart[1]['nextturn']]
            elif sign == '/':
                cartsdf.loc[i,'nextmove'] = 'left'
            elif sign == '\\':
                cartsdf.loc[i,'nextmove'] = 'right'
        elif cart[1]['nextmove'] == 'right':
            cartsdf.loc[i,'column'] += 1
            newcoords = (cartsdf.loc[i,'row'],cartsdf.loc[i,'column'])
            if newcoords in list(cartsdf['coords'].values):
                print('First crash happend at {}'.format(newcoords))
                return newcoords
            cartsdf.at[i,'coords'] = newcoords
            sign = startmap.iloc[cartsdf.loc[i,'row'],cartsdf.loc[i,'column']]
            if sign == '+':
                if cart[1]['nextturn'] == 'straight':
                    pass
                elif cart[1]['nextturn'] == 'left':
                    cartsdf.loc[i,'nextmove'] = trnle[cart[1]['nextmove']]
                elif cart[1]['nextturn'] == 'right':
                    cartsdf.loc[i,'nextmove'] = trnri[cart[1]['nextmove']]
                cartsdf.loc[i,'nextturn'] = nxttrn[cart[1]['nextturn']]
            elif sign == '/':
                cartsdf.loc[i,'nextmove'] = 'up'
            elif sign == '\\':
                cartsdf.loc[i,'nextmove'] = 'down'
        elif cart[1]['nextmove'] == 'left':
            cartsdf.loc[i,'column'] -= 1
            newcoords = (cartsdf.loc[i,'row'],cartsdf.loc[i,'column'])
            if newcoords in list(cartsdf['coords'].values):
                print('First crash happend at {}'.format(newcoords))
                return newcoords
            cartsdf.at[i,'coords'] = newcoords
            sign = startmap.iloc[cartsdf.loc[i,'row'],cartsdf.loc[i,'column']]
            if sign == '+':
                if cart[1]['nextturn'] == 'straight':
                    pass
                elif cart[1]['nextturn'] == 'left':
                    cartsdf.loc[i,'nextmove'] = trnle[cart[1]['nextmove']]
                elif cart[1]['nextturn'] == 'right':
                    cartsdf.loc[i,'nextmove'] = trnri[cart[1]['nextmove']]
                cartsdf.loc[i,'nextturn'] = nxttrn[cart[1]['nextturn']]
            elif sign == '/':
                cartsdf.loc[i,'nextmove'] = 'down'
            elif sign == '\\':
                cartsdf.loc[i,'nextmove'] = 'up'
    return None
```

```python
startmap, cartsdf = getcarts(startdf)
```

```python
def printstatus(basemap, carts):
    replace = {'right': '>', 'left': '<', 'up': '^', 'down': 'v'}
    newmap = basemap.copy()
    for cart in carts.iterrows():
        r,c = cart[1]['coords']
        newmap.iloc[r,c] = replace[cart[1]['nextmove']]
    return newmap
```

```python
def lastcart(startmap, cartsdf):
    nxttrn = {'right': 'left', 'straight': 'right', 'left': 'straight'}
    trnle = {'up': 'left', 'left': 'down', 'down': 'right', 'right': 'up'}
    trnri = {'up': 'right', 'right': 'down', 'down': 'left', 'left': 'up'}
    toremove = []
    for cart in cartsdf.sort_values(['row','column']).iterrows():
        i = cart[0]
        if cart[1]['nextmove'] == 'up':
            newcoords = (cartsdf.loc[i,'row']-1,cartsdf.loc[i,'column'])
            if newcoords in list(cartsdf['coords'].values):
                row, col = newcoords
                toremove.append(i)
                cart2 = cartsdf[(cartsdf['row'] == row) & (cartsdf['column'] == col)].index[0]
                toremove.append(cart2)
                print('Crash! carts {} and {} will be removed'.format(i,cart2))
            else:
                cartsdf.loc[i,'row'] -= 1
                cartsdf.at[i,'coords'] = newcoords
                sign = startmap.iloc[cartsdf.loc[i,'row'],cartsdf.loc[i,'column']]
                if sign == '+':
                    if cart[1]['nextturn'] == 'straight':
                        pass
                    elif cart[1]['nextturn'] == 'left':
                        cartsdf.loc[i,'nextmove'] = trnle[cart[1]['nextmove']]
                    elif cart[1]['nextturn'] == 'right':
                        cartsdf.loc[i,'nextmove'] = trnri[cart[1]['nextmove']]
                    cartsdf.loc[i,'nextturn'] = nxttrn[cart[1]['nextturn']]
                elif sign == '/':
                    cartsdf.loc[i,'nextmove'] = 'right'
                elif sign == '\\':
                    cartsdf.loc[i,'nextmove'] = 'left'
        elif cart[1]['nextmove'] == 'down':
            newcoords = (cartsdf.loc[i,'row']+1,cartsdf.loc[i,'column'])
            if newcoords in list(cartsdf['coords'].values):
                row, col = newcoords
                toremove.append(i)
                cart2 = cartsdf[(cartsdf['row'] == row) & (cartsdf['column'] == col)].index[0]
                toremove.append(cart2)
                print('Crash! carts {} and {} will be removed'.format(i,cart2))
            else:
                cartsdf.loc[i,'row'] += 1
                cartsdf.at[i,'coords'] = newcoords
                sign = startmap.iloc[cartsdf.loc[i,'row'],cartsdf.loc[i,'column']]
                if sign == '+':
                    if cart[1]['nextturn'] == 'straight':
                        pass
                    elif cart[1]['nextturn'] == 'left':
                        cartsdf.loc[i,'nextmove'] = trnle[cart[1]['nextmove']]
                    elif cart[1]['nextturn'] == 'right':
                        cartsdf.loc[i,'nextmove'] = trnri[cart[1]['nextmove']]
                    cartsdf.loc[i,'nextturn'] = nxttrn[cart[1]['nextturn']]
                elif sign == '/':
                    cartsdf.loc[i,'nextmove'] = 'left'
                elif sign == '\\':
                    cartsdf.loc[i,'nextmove'] = 'right'
        elif cart[1]['nextmove'] == 'right':
            newcoords = (cartsdf.loc[i,'row'],cartsdf.loc[i,'column']+1)
            if newcoords in list(cartsdf['coords'].values):
                row, col = newcoords
                toremove.append(i)
                cart2 = cartsdf[(cartsdf['row'] == row) & (cartsdf['column'] == col)].index[0]
                toremove.append(cart2)
                print('Crash! carts {} and {} will be removed'.format(i,cart2))
            else:
                cartsdf.loc[i,'column'] += 1
                cartsdf.at[i,'coords'] = newcoords
                sign = startmap.iloc[cartsdf.loc[i,'row'],cartsdf.loc[i,'column']]
                if sign == '+':
                    if cart[1]['nextturn'] == 'straight':
                        pass
                    elif cart[1]['nextturn'] == 'left':
                        cartsdf.loc[i,'nextmove'] = trnle[cart[1]['nextmove']]
                    elif cart[1]['nextturn'] == 'right':
                        cartsdf.loc[i,'nextmove'] = trnri[cart[1]['nextmove']]
                    cartsdf.loc[i,'nextturn'] = nxttrn[cart[1]['nextturn']]
                elif sign == '/':
                    cartsdf.loc[i,'nextmove'] = 'up'
                elif sign == '\\':
                    cartsdf.loc[i,'nextmove'] = 'down'
        elif cart[1]['nextmove'] == 'left':
            newcoords = (cartsdf.loc[i,'row'],cartsdf.loc[i,'column']-1)
            if newcoords in list(cartsdf['coords'].values):
                row, col = newcoords
                toremove.append(i)
                cart2 = cartsdf[(cartsdf['row'] == row) & (cartsdf['column'] == col)].index[0]
                toremove.append(cart2)
                print('Crash! carts {} and {} will be removed'.format(i,cart2))
            else:
                cartsdf.loc[i,'column'] -= 1
                cartsdf.at[i,'coords'] = newcoords
                sign = startmap.iloc[cartsdf.loc[i,'row'],cartsdf.loc[i,'column']]
                if sign == '+':
                    if cart[1]['nextturn'] == 'straight':
                        pass
                    elif cart[1]['nextturn'] == 'left':
                        cartsdf.loc[i,'nextmove'] = trnle[cart[1]['nextmove']]
                    elif cart[1]['nextturn'] == 'right':
                        cartsdf.loc[i,'nextmove'] = trnri[cart[1]['nextmove']]
                    cartsdf.loc[i,'nextturn'] = nxttrn[cart[1]['nextturn']]
                elif sign == '/':
                    cartsdf.loc[i,'nextmove'] = 'down'
                elif sign == '\\':
                    cartsdf.loc[i,'nextmove'] = 'up'
    cartsdf.drop(toremove, inplace=True)
    if len(cartsdf) == 1:
        print('Last cart is at position {}'.format(cartsdf['coords']))
        return cartsdf['coords'].item()
    return None
```
